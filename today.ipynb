```python
from castlib import Event, Cast, Prop, Reporter
from datetime import date
from dateutil.relativedelta import relativedelta
```

```python
d = date
c = (Cast(balance=2185 + 22000, start=date.today() + relativedelta(days=1))
     .end_after(years=1)
     .report_with(Reporter)
     .set_min_cash(5000)
     .at_apr(10)
     .add_semi_monthly(Event(name="Saleae Paycheck", amount=5400, date=d(2020, 2, 15)))
     
     .add_monthly(Event(name="Rent", amount=-3900, date=d(2020, 2, 1)), end=d(2020, 6, 1))
     .add_monthly(Event(name="Rent", amount=-3000, date=d(2020, 7, 1)))
     .add_monthly(Event(name="Utils", amount=-150, date=d(2020, 2, 15)))
     
     .add_weekly(Event(name="Food", amount=-100, date=d(2020, 2, 2)))
     .add_weekly(Event(name="Gas", amount=-60, date=d(2020, 2, 2)))
     
     .add_monthly(Event(name="Carmax", amount=-910, date=d(2020, 2, 18)), end=d(2020, 12, 31))
     .add_monthly(Event(name="Progressive", amount=-310, date=d(2020, 2, 18)))
     
     .add_monthly(Event(name="Sprint", amount=-450, date=d(2020, 2, 18)), end=d(2020, 6, 1))
     .add_monthly(Event(name="Sprint", amount=-200, date=d(2020, 6, 1)))
    )

TripToEurope = Prop([
    Event(name="Flights", amount=-1200, date=d(2020, 6, 1)),
    Event(name="Trip", amount=-1500, date=d(2020, 7, 1))
], "TripToEurope")

MapsProjectPayout = Prop([
    Event(name="First 50%", amount=15000, date=d(2020, 5, 1)),
    Event(name="Second 50%", amount=15000, date=d(2020, 9, 1))
], "MapsProjectPayout")

c.report.running_balance(props=[TripToEurope, MapsProjectPayout]);
```
