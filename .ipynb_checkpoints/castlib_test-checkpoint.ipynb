```python
import datetime
from dateutil.relativedelta import relativedelta
import time
import json
import math
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker

from castlib.cast import Cast
from castlib.reporter import Reporter
from castlib.prop import Prop
from castlib.event import Event
from castlib.util import plot_casts
```

```python
d = datetime.date

def AccountProp(name, amount):
        return Event(name=f"cashout {name}", amount=amount, date=d(2020, 3, 1))

month_of_severance = 15400
num_months = 0.5
severance = month_of_severance * num_months

# 401K
cashout_401k = False
Guideline = AccountProp("401k", 8500 * 0.7)

# Investments
Schwab = AccountProp("SchwabPortfolios", 7000)
Robinhood = AccountProp("Robinhood", 1000)
Wealthfront = AccountProp("Wealthfront", 10000)

# Unemployment
unempoloyemnt_event = Event(name="Unemployment", amount=450, date=d(2020, 3, 10))
unempoloyemnt_event.end = d(2020, 3, 10) + relativedelta(weeks=26)
Unemployment = Prop(unempoloyemnt_event.weekly, name="Unemployment")

# Cash
personal_capital_cash = 7000 + 14000

desk = 400
chair = 599
monitor = 500

Desk = Prop([Event(name="Desk", amount=-(desk + (chair * 2) + monitor), date=d(2020, 3, 1))], name="Desk")

c = (Cast(balance=personal_capital_cash)
     .end_after(months=14)
     .report_with(Reporter)
     
     .add_event(Event(name="Severance", amount=severance, date=d(2020, 3, 1)))
     .add_event(Event(name="Kennedy Flats", amount=1000, date=d(2020, 3, 1)))

     # with nora
     .add_event(Event(name="Rent", amount=-3900, date=d(2020, 2, 1), end=d(2020, 7, 31)).monthly)
     
     # after nora
     .add_event(Event(name="Rent", amount=-2000, date=d(2020, 8, 1)).monthly)

     .add_event(Event(name="Carmax", amount=-910, date=d(2020, 5, 16)).monthly)
     .add_event(Event(name="Progressive", amount=-310, date=d(2020, 2, 10)).monthly)
     .add_event(Event(name="Utilities", amount=-250, date=d(2020, 3, 15)).monthly)
     .add_event(Event(name="Sprint", amount=-400, date=d(2020, 3, 15)).monthly)
     .add_event(Event(name="Medical", amount=-100, date=d(2020, 3, 15)).monthly)

     .add_event(Event(name="Food", amount=-150, date=d(2020, 2, 2)).weekly)
     .add_event(Event(name="Gas", amount=-35, date=d(2020, 2, 2)).weekly)
     
     .add_event([Schwab, Robinhood])
     .add_event(Desk.events)
     .add_event(Unemployment.events)

     .at_apr(10)
     .save(f"/Users/jacobsansbury/Casts/fired-{d.today()}.json")
    )

print("Best Case")
months_til_broke = ((c.first_day_under(0) - d.today()).days / 30)
print(f"{c.first_day_under(0)} - {str(months_til_broke)[:4]} months")

print("\nWorst Case")
c.add_event([Guideline, Wealthfront])
months_til_broke = ((c.first_day_under(0) - d.today()).days / 30)
print(f"{c.first_day_under(0)} - {str(months_til_broke)[:4]} months")

c.report.running_balance();
```

```python
b = Cast.fromfile("/Users/jacobsansbury/Casts/fired-2020-02-25.json").report_with(Reporter)
```

```python
def plot_casts(cs):
    fig, ax = cs[0].report.figure()
    formatter = ticker.StrMethodFormatter('${x:,.0f}')
    ax.yaxis.set_major_formatter(formatter)
    
    for i, c in enumerate(cs):
        dates, balances = c.running_balance(split=True)
        date_times = [time.mktime(d.timetuple()) for d in dates]
        plt.plot(date_times, balances, label=f'Cast {i}')
    
plot_casts([c, b])
```
